```
(
  df
 .groupby('param_marka-pojazdu')['price_value']
 .agg([np.median, np.mean, np.size])
 .sort_values(by='mean', ascending=False)
 .head(25)
) .plot(kind='bar', figsize=(18,5))
```

```
(
  df
 .groupby('param_marka-pojazdu')['price_value']
 .agg([np.median, np.mean, np.size])
 .sort_values(by='mean', ascending=False)
 .head(40)
) .plot(kind='bar', figsize=(18,5), subplots=True)
```

```
(
  df
 .groupby('param_marka-pojazdu')['price_value']
 .agg([np.median, np.mean, np.size])
 .sort_values(by='size', ascending=False)
 .head(50)
) .plot(kind='bar', figsize=(18,5), subplots=True)
```

```
def group_and_barplot(feat_grupby, feat_agg='price_value', agg_funcs=[np.median, np.mean, np.size],feat_sort='mean', top=50, subplots=True):
  return(
  df
 .groupby(feat_grupby)[feat_agg]
 .agg(agg_funcs)
 .sort_values(by=feat_sort, ascending=False)
 .head(top)
) .plot(kind='bar', figsize=(18,5), subplots=subplots)
```

```
group_and_barplot('param_marka-pojazdu');
```

```
group_and_barplot('param_kraj-pochodzenia');
```

```
group_and_barplot('param_kraj-pochodzenia', feat_sort='size');
```

```
group_and_barplot('param_marka-pojazdu');
```

```
group_and_barplot('param_kolor', feat_sort='size');
```

```
group_and_barplot('param_pojemność-skokowa', feat_sort='size');
```
